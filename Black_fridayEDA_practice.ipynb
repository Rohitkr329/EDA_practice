```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#combining two files using the append fx
df_final = df1.append(df)
df_final.head()
```

```python
#dropping user_id as it's of no use
df_final.drop(['User_ID'], axis =1 , inplace = True)
df_final.head()
```

```python
#mappping genders using map fx
df_final['Gender']  = df_final['Gender'].map({'F':1, 'M':0})
```

```python
df_final.head()
```

```python
#unique values in age columns

df_final.Age.unique()
```

```python
#mapping the ages to numbers

df_final['Age']=df_final['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
```

```python
df_final.head()
```

```python
df.head()
```

```python
#getting dummies and dropping the first
df_city=pd.get_dummies(df_final['City_Category'],drop_first=True)
```

```python
df_city
```

```python
df_final=pd.concat([df_final,df_city],axis=1)
df_final.head()
```

```python
#checking the rows and no of columns
df_final.shape
```

```python
##drop City Category Feature
df_final.drop('City_Category',axis=1,inplace=True)
```

```python
df_final.head()
```

```python
## Missing Values
df_final.isnull().sum()
```

```python
## Focus on replacing missing values
df_final['Product_Category_2'].unique()
```

```python
df_final['Product_Category_2'].value_counts()
```

```python
#finding mode val
df_final['Product_Category_2'].mode()[0]
```

```python
## Replace the missing values with mode

df_final['Product_Category_2']=df_final['Product_Category_2'].fillna(df_final['Product_Category_2'].mode()[0])
```

```python
# again checking for the null values in the product category 2
df_final['Product_Category_2'].isnull().sum()
```

```python
#checking for the missing values in the product cat 3

df_final['Product_Category_3'].isnull().sum()
```

```python
#we can see there is a "+" sign after the 4 which makes the whole column's dtype as object hence replacing the same with a null

df_final['Stay_In_Current_City_Years']=df_final['Stay_In_Current_City_Years'].str.replace('+','')
```

```python
## now let's convert the uint8 to int
df_final['B']=df_final['B'].astype(int)
df_final['C']=df_final['C'].astype(int)
```
